```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
d = pickle.load(open("/home/rafael/git/face_detection_clustering/video/gal_gadot_pam/features.pkl",'rb'))
frames = list(d.keys())
```

```python
scores = []
for f1,f2 in zip(frames[:-1],frames[1:]):
    if f2-f1!=1:
        print('frames n√£o sequenciais')
    scores.append(np.dot(d[f1],d[f2]))
scores = np.array(scores)
```

```python
x = np.arange(len(scores))/23.98
plt.plot(x,scores)
plt.ylim(0.5,1)
```
